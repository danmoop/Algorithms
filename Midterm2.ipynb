```python
import tensorflow as tf
import sklearn
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import numpy as np

# Function to process train/test labels
def process(input):
  result = []
  for i in range(0, len(input)):
    arr = [0, 0, 0]
    arr[input[i]] = 1
    result.append(arr)
  return np.array(result)

# Loading the dataset
dataset = load_iris(return_X_y=False, as_frame=False)
data = dataset.data

# Splitting the data into 70% (train data) 30% (test data)
train_data, test_data, train_labels, test_labels = train_test_split(dataset.data, dataset.target, random_state=42, test_size=0.3)

# Converting labels from 1D to categorical label (e.g [1, 0, 0] and [0, 1, 0])
train_labels_processed = process(train_labels)
test_labels_processed = process(test_labels)

# Creating a model
model = tf.keras.Sequential()

for i in range(0, 8):
  model.add(tf.keras.layers.Dense(units=12, activation='sigmoid'))

model.add(tf.keras.layers.Dense(units=3, activation='softmax'))
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])

# Training the model
history = model.fit(train_data, train_labels_processed, epochs=1000, verbose=0, validation_split=0.1)

# Evaluating based on test (unseen) data
test_loss, test_accuracy = model.evaluate(test_data, test_labels_processed)
print('Test loss: {0:.2f}. Test accuracy: {1:.2f}%'.format(test_loss, test_accuracy*100.))

c = ['red', 'green', 'blue']
colors = []

for i in range(0, len(test_data)):
  t = np.array([test_data[i]])
  prediction = model.predict(t)
  max_index = np.argmax(prediction)
  colors.append(c[max_index])

sepalLengths = []
sepalWidths = []
petalLengths = []
petalWidths = []
x = []

for i in range(0, len(test_data)):
  x.append(i)
  sepalLengths.append(test_data[i][0])
  sepalWidths.append(test_data[i][1])
  petalLengths.append(test_data[i][2])
  petalWidths.append(test_data[i][3])


plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

fig, axs = plt.subplots(2, 2)

axs[0, 0].scatter(x, sepalLengths, c=colors)
axs[0, 0].set_title('sepal length (cm)')

axs[0, 1].scatter(x, sepalWidths, c=colors)
axs[0, 1].set_title('sepal width (cm)')

axs[1, 0].scatter(x, petalLengths, c=colors)
axs[1, 0].set_title('petal length (cm)')

axs[1, 1].scatter(x, petalWidths, c=colors)
axs[1, 1].set_title('petal width (cm)')

plt.tight_layout()
plt.show()
```
